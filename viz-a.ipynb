```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## As shown Below We have a table of data considering the amount of tips for each day, so let's graph it to see the relation between the day and the amount of tip

```python
sns.get_dataset_names()
tips = sns.load_dataset('tips')
tips.head(30)
```

```python

average_tip_day = tips.groupby("day", as_index=False).tip.mean()
sns.barplot(x='day',y="tip",data=average_tip_day)
```

## As we can see in the barplot we have a pattern that people tend to give tips more on sundays than other days.
## As people are more generous on sundays then we can make offers for specific meals on every sunday as to make them buy more. 

## Let's find out if their average total bill is it affected by specific time;

```python
average_bill_time = tips.groupby("time", as_index=False).total_bill.mean()
sns.barplot(x="time",y="total_bill",data=average_bill_time)
```

## AS we can see People tend to pay more on meals at dinner time so I would make offer speically on dinners meals to push them to buy more.

## Conclusions:

1- We can make special offers on sunday .

2- We can make special offers on evening meals.
